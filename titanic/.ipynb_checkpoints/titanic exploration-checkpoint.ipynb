```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.nonparametric.kde import KDEUnivariate
from statsmodels.nonparametric import smoothers_lowess
from pandas import Series, DataFrame
from patsy import dmatrices
from sklearn import datasets, svm
import os
```

```python
df=pd.read_csv('titanic_data.csv')
df.head()
```

```python
df.isnull().sum()
```

Cabin doesn't add much, so can ignore, will clean up and drop NAs

```python
df2=df.drop(['Cabin'], axis=1)
#remove NAs
df2 = df2.dropna()
```

Plot out the data to have a look at it

```python
#specify parameters of graphs
fig = plt.figure(figsize=(18,6), dpi=1600)
alpha = alpha_scatterplot = 0.2
alpha_bar_chart = 0.55

#plot multiple different graphs together
ax1 = plt.subplot2grid((2,3),(0,0)) # this denotes placement of the graph in the output, 2 rows and 3 columns of graphs
#plot a bar graph of those who survived vs those who didn't
df2.Survived.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
# this nicely sets the margins in matplotlib to deal with a recent bug 1.3.1
ax1.set_xlim(-1,2)
#title the graph
plt.title('Distribution of Survival, 1 = Survived')

plt.subplot2grid((2,3),(0,1)) #saying this will be first row 2nd column
plt.scatter(df2.Survived, df2.Age,alpha=alpha_scatterplot)
plt.ylabel('Age') #label y axis
#format grid lines w b=True
plt.grid(b=True, which = 'major', axis = 'y')
plt.title('Survival by Age')

ax3 = plt.subplot2grid((2,3),(0,2))
df2.Pclass.value_counts().plot(kind='barh', alpha=alpha_bar_chart)
ax3.set_ylim(-1, len(df2.Pclass.value_counts()))
plt.title('Class Distribution')

plt.subplot2grid((2,3),(1,0), colspan=2) #colspan =2 allows the wider graph to spread 2 columns
# plots a kernel density estimate of the passangers's age by class
df2.Age[df2.Pclass==1].plot(kind = 'kde')
df2.Age[df2.Pclass==2].plot(kind = 'kde')
df2.Age[df2.Pclass==3].plot(kind = 'kde')
plt.xlabel("Age") #label x axis
plt.title('Age Distribution within Classes') #title the graph
plt.legend(('1st Class','2nd Class','3rd Class'), loc='best')

ax5 = plt.subplot2grid((2,3),(1,2))
df2.Embarked.value_counts().plot(kind='bar', alpha = alpha_bar_chart)
ax5.set_xlim(-1, len(df2.Embarked.value_counts())) #specifies x parameters
plt.title('Passengers per Boarding Location')
```
